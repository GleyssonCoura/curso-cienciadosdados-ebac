```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sinasc = pd.read_csv('SINASC_RO_2019.csv')
sinasc.head()
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
sinasc = sinasc[sinasc['APGAR5'].notna()]
sinasc.loc[sinasc['APGAR5'] >= 8, 'APGAR5_CLASSIFICACAO'] = 'normal'
sinasc.loc[(sinasc['APGAR5'] >= 6) & (sinasc['APGAR5'] < 8), 'APGAR5_CLASSIFICACAO'] = 'asfixia_leve'
sinasc.loc[(sinasc['APGAR5'] >= 4) & (sinasc['APGAR5'] < 6), 'APGAR5_CLASSIFICACAO'] = 'asfixia_moderada'
sinasc.loc[sinasc['APGAR5'] < 4, 'APGAR5_CLASSIFICACAO'] = 'asfixia_grave'
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
plt.rc('figure', figsize=(10, 10))
fig, axes = plt.subplots(2, 3)

sns.countplot(ax = axes[0,0], x='APGAR5_CLASSIFICACAO',hue='SEXO',
              data=sinasc[sinasc['GESTACAO']=='Menos de 22 semanas']).set_xticklabels(sns.countplot(ax = axes[0,0], x='APGAR5_CLASSIFICACAO',hue='SEXO',
                                                                                                    data=sinasc[sinasc['GESTACAO']=='Menos de 22 semanas']).get_xticklabels(),rotation=45, horizontalalignment='right')
sns.countplot(ax = axes[0,1], x='APGAR5_CLASSIFICACAO',hue='SEXO',
              data=sinasc[sinasc['GESTACAO']=='22 a 27 semanas']).set_xticklabels(sns.countplot(ax = axes[0,1], x='APGAR5_CLASSIFICACAO',hue='SEXO',
                                                                                                    data=sinasc[sinasc['GESTACAO']=='22 a 27 semanas']).get_xticklabels(),rotation=45, horizontalalignment='right')

sns.countplot(ax = axes[0,2], x='APGAR5_CLASSIFICACAO',hue='SEXO',
              data=sinasc[sinasc['GESTACAO']=='28 a 31 semanas']).set_xticklabels(sns.countplot(ax = axes[0,2], x='APGAR5_CLASSIFICACAO',hue='SEXO',
                                                                                                    data=sinasc[sinasc['GESTACAO']=='28 a 31 semanas']).get_xticklabels(),rotation=45, horizontalalignment='right')

sns.countplot(ax = axes[1,0], x='APGAR5_CLASSIFICACAO',hue='SEXO',
              data=sinasc[sinasc['GESTACAO']=='32 a 36 semanas']).set_xticklabels(sns.countplot(ax = axes[1,0], x='APGAR5_CLASSIFICACAO',hue='SEXO',
                                                                                                    data=sinasc[sinasc['GESTACAO']=='32 a 36 semanas']).get_xticklabels(),rotation=45, horizontalalignment='right')

sns.countplot(ax = axes[1,1], x='APGAR5_CLASSIFICACAO',hue='SEXO',
              data=sinasc[sinasc['GESTACAO']=='37 a 41 semanas']).set_xticklabels(sns.countplot(ax = axes[1,1], x='APGAR5_CLASSIFICACAO',hue='SEXO',
                                                                                                    data=sinasc[sinasc['GESTACAO']=='37 a 41 semanas']).get_xticklabels(),rotation=45, horizontalalignment='right')

sns.countplot(ax = axes[1,2], x='APGAR5_CLASSIFICACAO',hue='SEXO',
              data=sinasc[sinasc['GESTACAO']=='42 semanas e mais']).set_xticklabels(sns.countplot(ax = axes[1,2], x='APGAR5_CLASSIFICACAO',hue='SEXO',
                                                                                                    data=sinasc[sinasc['GESTACAO']=='42 semanas e mais']).get_xticklabels(),rotation=45, horizontalalignment='right')

plt.subplots_adjust(wspace=0.5, hspace=0.5)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
# Tabela cruzada
freq = pd.crosstab(sinasc['GESTACAO'], sinasc['APGAR5'])
freq
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
freq_pctlin = freq.div(freq.sum(axis = 1), axis = 0)
freq_pctlin
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
freq_pctlin.plot.bar()
```

Considerando que nascimento prematuro significa um bebê menos "pronto", percebe-se que quanto menos semanas de gestação, maior a ocorrência de graus de asfixia nos recém nascidos (APGAR < 8)

Como APGAR5 é uma variável discreta, as melhores visualizações são histogramas e boxplots.

```python
sns.boxplot(data = sinasc, y="PESO", x="APGAR5", orient="v")
```

O boxplot acima traz uma informação interessante mostrando a tendência de peso mediano menor dos bebês conforme aumenta a severidade do grau de asfixia. Contudo, o resultado de APGAR = 1 é intrigante.

```python
sns.boxplot(data = sinasc, y='PESO', x='APGAR1', orient="v")
```

Se comparar APGAR1 ao peso, temos uma tendência semelhante em relação à comparação de APGAR5 e PESO.

```python
sns.boxplot(data = sinasc[sinasc['APGAR5']==1], y='PESO', x='APGAR1', orient="v")
```

Dentre os bebês que apresentaram APGAR5 = 1, ao analisar APGAR1 as mais leves seguem a tendência vista no primeiro gráfico entre APGAR5 e Peso, porém as mais pesadas não. Inclusive há um resultado contrastante para essas crianças mais pesadas, indicando alguma outra ocorrência que o peso pode ter influência.

```python
sns.boxplot(data = sinasc[(sinasc['APGAR5']==1)&(sinasc['APGAR1']>1)], y='PESO', x='SEXO', orient="v")
```

Após comparar as demais variáveis (SEXO, PARTO, GRAVIDEZ, GESTACAO, IDANOMAL) na condição de APGAR5 = 1 e APGAR1 >1 para tentar entender alguma relação entre as ocorrências de asfixia grave em crianças após o 5º minuto e um peso "normal" (> 3kg), não foi possível encontrar relação com as variáveis disponíveis.

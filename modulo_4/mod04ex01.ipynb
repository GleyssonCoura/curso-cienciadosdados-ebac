## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1) Seu código aqui
acertos_mat = (df_mat['Qt_acertos'].sum())/24
print(f'O percentual de acerto de matemática foi de: {acertos_mat:,.2f} %')

acertos_por = (df_por['Qt_acertos'].sum())/18
print(f'O percentual de acerto de português foi de: {acertos_por:,.2f} %')

acertos_geo = (df_geo['Qt_acertos'].sum())/8
print(f'O percentual de acerto de geografia foi de: {acertos_geo:,.2f} %')

acertos_ing = (df_ing['Qt_acertos'].sum())/8
print(f'O percentual de acerto de inglês foi de: {acertos_ing:,.2f} %')

acertos_his = (df_his['Qt_acertos'].sum())/8
print(f'O percentual de acerto de história foi de: {acertos_his:,.2f} %')

acertos_fis = (df_fis['Qt_acertos'].sum())/12
print(f'O percentual de acerto de física foi de: {acertos_fis:,.2f} %')

acertos_qui = (df_qui['Qt_acertos'].sum())/12
print(f'O percentual de acerto de química foi de: {acertos_qui:,.2f} %')
````

````python
# 2) Seu código aqui
total_acertos = df_mat + df_por + df_geo + df_ing + df_his + df_fis + df_qui

total_acertos['Aluno'] = np.arange(1,101,1)
total_acertos[['Aluno','Qt_acertos']]
````

````python
# 3) Seu código aqui
total_acertos['Percentual_geral'] = round((total_acertos['Qt_acertos']/(24+18+8+8+8+12+12))*100,2)
total_acertos[['Aluno','Percentual_geral']]
````

````python
# 4) Seu código aqui
aprovados = total_acertos[total_acertos['Qt_acertos'] > 45]
quantidade_aprovados = len(aprovados['Aluno'])
quantidade_aprovados
````

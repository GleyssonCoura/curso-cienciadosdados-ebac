````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_pickle('base_nfp.pkl')
df.info()
````

## Criando variveis

**Variveis temporais**:  possvel que haja alguma regra de liberao de crditos de acordo com o dia do ms, o dia da semana, fim de semana (sim ou no), ou at alteraes temporais da regra. Vamos criar as seguintes variveis para testar estas hipteses:

- Dia da semana
- Dummy para fim de semana
- Dia do ms
- Ms
- Trimestre
- Ano (j na base)

**Valor da nota**: Vamos criar tambm algumas variveis para resumir o valor da nota, e posteriormente criar uma intuio sobre categorizaes de variveis contnuas.

- Crie categorizaes da varivel ```Valor NF```, por quants, em 5, 10, 20 e 50 categorias.

````python
dia_semana = {0:'domingo',1:'segunda',2:'terca',3:'quarta',4:'quinta',5:'sexta',6:'sabado'}
df['dia_semana'] = df['Data Emisso'].dt.weekday.map(dia_semana)
df['fim_de_semana'] = [1 if x == 'domingo' else 1 if x == 'sabado' else 0 for x in df['dia_semana']]
df['dia_mes'] = df['Data Emisso'].dt.day
df['mes'] = df['Data Emisso'].dt.month
trimestre = {1:1,2:1,3:1,4:2,5:2,6:2,7:3,8:3,9:3,10:4,11:4,12:4}
df['trimestre'] = df['mes'].map(trimestre)
df['5_categorias'] = pd.qcut(df['Valor NF'],5)
df['10_categorias'] = pd.qcut(df['Valor NF'],10)
df['20_categorias'] = pd.qcut(df['Valor NF'],20)
df['50_categorias'] = pd.qcut(df['Valor NF'],50)
````

## Calcule o IV de todas essas variveis

Sugesto: Construa um metadados e tente utilizar uma funo como a que foi utilizada durante a aula. Voc vai conseguir fazer esse clculo para as variveis uma a uma, mas em casos reais em que vamos mais alm na criao de variveis, podemos ter centenas, ou at milhares de variveis, esses clculos *precisam* ser automticos nesses casos.

````python
metadados = pd.DataFrame({'dtype':df.dtypes})
metadados['papel'] = 'covariavel'
metadados.loc['flag_credito','papel'] = 'resposta'
metadados['distintos'] = df.nunique()

def IV(variavel, resposta):
    tab = pd.crosstab(variavel, resposta, margins=True, margins_name='total')

    rtulo_evento = tab.columns[0]
    rtulo_nao_evento = tab.columns[1]

    tab['pct_evento'] = tab[rtulo_evento]/tab.loc['total',rtulo_evento]
    tab['pct_nao_evento'] = tab[rtulo_nao_evento]/tab.loc['total',rtulo_nao_evento]
    tab['woe'] = np.log(tab.pct_evento/tab.pct_nao_evento)
    tab['iv_parcial'] = (tab.pct_evento - tab.pct_nao_evento)*tab.woe
    return tab['iv_parcial'].sum()

for var in metadados[metadados.papel=='covariavel'].index:
    if  (metadados.loc[var, 'distintos']>5):
        metadados.loc[var, 'IV'] = IV(df[var], df.flag_credito)
    else: 
        metadados.loc[var, 'IV'] = IV(df[var], df.flag_credito)

metadados
````

## Conclua

- Quais variveis parecem apresentar maior poder de discriminao segundo o IV?
- Quando categorizamos variveis contnuas parece haver alguma relao entre o nmero de categorias e o IV?

````python
metadados.sort_values(by='IV',ascending = False)
````

- A varivel 'categoria'  a que apresenta maior poder de discriminao segundo o IV
- Percebe-se que quanto maior a quantidade de categorias, mais  necessrio que faa menores categorizaes, a fim de conseguir calcular um IV relevante

## Avalie o WOE e o IV no tempo

Avalie o WOE e o IV no tempo da varivel ```categoria```.

````python
tab = pd.DataFrame(df.groupby(['Data Emisso','categoria'])['flag_credito'].sum())
tab['total'] = pd.DataFrame(df.groupby(['Data Emisso','categoria'])['flag_credito'].count())
tab.rename(columns={'flag_credito':'credito_positivo'},inplace=True)
tab['diferenca'] = tab.total - tab.credito_positivo
tab['tx_evento'] = tab.credito_positivo/tab.total
tab['pct_evento'] = tab['credito_positivo']/tab['credito_positivo'].groupby(level=0).sum()
tab['pct_nao_evento'] = tab['diferenca']/tab['diferenca'].groupby(level=0).sum()
tab['WOE'] = np.log(tab.pct_evento/tab.pct_nao_evento)
````

````python
tab_reduzido = tab[(tab['WOE']>0)&(tab['diferenca']!=0)]
tab_reduzido
````

````python
fig, ax = plt.subplots()
variavel='categoria'

# Comando principal do grfico
sns.pointplot(data=tab_reduzido.reset_index(), y = 'WOE', x = 'Data Emisso', hue=variavel, ax=ax);

# Rotacionar os rtulos do eixo X
labs = tab_reduzido.index.get_level_values(level=0).unique()
labs2 = [x.to_period('M') for x in labs]

ax.set_xticklabels(labs2,rotation=30, fontsize='small')
ax.legend(bbox_to_anchor=(.65, -.2), ncol=2)

#limites do grfico - mostrando 1 e -1 para ter referncia
ymax = max(1.2, max(tab_reduzido.WOE))
ymin = min(-1.2, min(tab_reduzido.WOE))
ax.set_ylim([ymin, ymax])

# Linhas de referncia
ax.axhline(1, linestyle=':'  , color='red', alpha=.5)
ax.axhline(0, linestyle='--' , color='grey', alpha=.5)
ax.axhline(-1, linestyle=':' , color='red', alpha=.5)

# Ttulos
ax.set_title(f'WOE ao longo do tempo para varivel {variavel.capitalize()}')
ax.set_xlabel('Tempo', loc='right');
````

````python
iv_categoria = ((tab_reduzido.pct_evento - tab_reduzido.pct_nao_evento)*tab_reduzido.WOE).groupby(level=0).sum();
iv_categoria = iv_categoria.to_frame()
iv_categoria_reduzido = iv_categoria[iv_categoria[0]<=0.6]
````

````python
def plot_iv(IV):
    fig, ax = plt.subplots()
    x = IV.index
    a = [.02]*len(IV.index)
    b = [.1]*len(IV.index)
    c = [.3]*len(IV.index)
    d = [.5]*len(IV.index)
    e = [.6]*len(IV.index)

    ax.fill_between(IV.index, a, color='grey', alpha=.2)
    ax.fill_between(IV.index, a, b, color='orange', alpha=.1)
    ax.fill_between(IV.index, b, c, color='green', alpha=.1)
    ax.fill_between(IV.index, c, d, color='blue', alpha=.1)
    ax.fill_between(IV.index, d, e, color='purple', alpha=.1)
    ax.plot(IV)

    labs = IV.index.get_level_values(level=0).unique()
    labs2 = [x.to_period('M') for x in labs]

    ax.set_xticks(labs)
    ax.set_xticklabels(labs2,rotation=30, fontsize='small')

    ax.set_title("Information Value de Sexo ao longo das safras")
    ax.set_ylabel("Information Value")
    ax.set_xlabel("Safra")
    
plot_iv(iv_categoria_reduzido)
````

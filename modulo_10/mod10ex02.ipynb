````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
dados = pd.read_csv('previsao_de_renda.csv')
````

````python
df1 = dados[['renda','posse_de_imovel','posse_de_veiculo']].copy()
````

````python
df1.info()
````

````python
sns.pointplot(x="posse_de_imovel", y="renda",data=df1, dodge=True, ci = 90)
````

````python
sns.pointplot(x="posse_de_veiculo", y="renda",data=df1, dodge=True, ci = 90)
````

Como visto, ambas variáveis indicam uma boa predição de renda, haja vista a distância significativa entre as médias.

Assim, tanto a posse de imóvel, quanto a posse de veículo são boas preditoras de uma renda média maior, se comparada a não posse imóvel ou veículo.

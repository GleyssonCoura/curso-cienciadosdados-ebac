1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
dados = pd.read_csv('previsao_de_renda.csv')
````

````python
dados.info()
````

````python
dados_qualitativos = dados.select_dtypes(include=['object'])
dados_qualitativos.info()
````

````python
graf = sns.countplot(x='data_ref',hue='sexo',data=dados_qualitativos)
tick_labs = dados_qualitativos['data_ref'].unique()
ticks = graf.set_xticks(list(range(dados_qualitativos['data_ref'].nunique())))
labels = graf.set_xticklabels(tick_labs, rotation=45)
````

````python
graf = sns.countplot(x='data_ref',hue='tipo_renda',data=dados_qualitativos)
tick_labs = dados_qualitativos['data_ref'].unique()
ticks = graf.set_xticks(list(range(dados_qualitativos['data_ref'].nunique())))
labels = graf.set_xticklabels(tick_labs, rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
graf = sns.countplot(x='data_ref',hue='educacao',data=dados_qualitativos)
tick_labs = dados_qualitativos['data_ref'].unique()
ticks = graf.set_xticks(list(range(dados_qualitativos['data_ref'].nunique())))
labels = graf.set_xticklabels(tick_labs, rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
graf = sns.countplot(x='data_ref',hue='estado_civil',data=dados_qualitativos)
tick_labs = dados_qualitativos['data_ref'].unique()
ticks = graf.set_xticks(list(range(dados_qualitativos['data_ref'].nunique())))
labels = graf.set_xticklabels(tick_labs, rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
graf = sns.countplot(x='data_ref',hue='tipo_residencia',data=dados_qualitativos)
tick_labs = dados_qualitativos['data_ref'].unique()
ticks = graf.set_xticks(list(range(dados_qualitativos['data_ref'].nunique())))
labels = graf.set_xticklabels(tick_labs, rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

Entre as variáveis qualitativas existentes na base, aquela que demonstrou ser interessante aprofundar a análise em função da mudança no perfil após determinado momento é a variável 'educação'.

2) Para cada variável quantitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
graf = sns.pointplot(x="data_ref",y="qtd_filhos",hue='sexo',data=dados,dodge=True,ci=95)
tick_labs = dados['data_ref'].unique()
ticks = graf.set_xticks(list(range(dados['data_ref'].nunique())))
labels = graf.set_xticklabels(tick_labs, rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
graf = sns.pointplot(x="data_ref",y="idade",hue='sexo',data=dados,dodge=True,ci=95)
tick_labs = dados['data_ref'].unique()
ticks = graf.set_xticks(list(range(dados['data_ref'].nunique())))
labels = graf.set_xticklabels(tick_labs, rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
dados_sem_na = dados.dropna()

graf = sns.pointplot(x="data_ref",y="tempo_emprego",hue='sexo',data=dados,dodge=True,ci=95)
tick_labs = dados_sem_na['data_ref'].unique()
ticks = graf.set_xticks(list(range(dados_sem_na['data_ref'].nunique())))
labels = graf.set_xticklabels(tick_labs, rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
graf = sns.pointplot(x="data_ref",y="renda",hue='sexo',data=dados,dodge=True,ci=95)
tick_labs = dados['data_ref'].unique()
ticks = graf.set_xticks(list(range(dados['data_ref'].nunique())))
labels = graf.set_xticklabels(tick_labs, rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

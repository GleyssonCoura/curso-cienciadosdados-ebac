````python
import pandas as pd
import numpy as np

from pandas_datareader import data as pdr
import yfinance as yfin
yfin.pdr_override()

import plotly.express as px
from plotly import graph_objects
from plotly.graph_objs import Layout
````

````python
df = pd.read_csv('dados_automotivos.csv')
df = df.drop(['Unnamed: 0'],axis=1)
````

#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
#1
df['media_movel_90'] = df['ajuste_fechamento'].rolling(90).mean()
````

````python
#2
df['desvio_padrao'] = df['ajuste_fechamento'].rolling(90).std()
````

````python
#3
fig = px.line(df, x="data", y="media_movel_90", color='ativo', template='plotly_white')
fig.show()
````

````python
#4
fig = px.line(df, x="data", y="desvio_padrao", color='ativo', template='plotly_white')
fig.show()
````

### 5
Os gráficos aponntam para resultados bem semelhantes, o que faz sentido ao se pensar que tanto a média móvel quanto o desvio padrão trazem um resultado de dispersão dos dados frente à média.

A diferença está na amplitude dos valores, uma vez que a média móvel cria uma maior suavização dos dados, principalmente em função do grande período de dias que é feita a avaliação da média.

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
df_reduzido = df.loc[(df['ativo']=='TSLA') & (df['data'] > '2022-11-01')]
````

````python
df.head()
````

````python
graph = {
    'x': df_reduzido.index,
    'open': df_reduzido.abertura,
    'close': df_reduzido.fechamento,
    'high': df_reduzido.alta,
    'low': df_reduzido.baixa,
    'type': 'candlestick',
    'name': 'TSLA',
    'showlegend': True
}
````

````python
layout = Layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)'
)

grafico = graph_objects.Figure(
    data = [graph],
    layout_title="Ações da Tesla",
    layout=layout
)
grafico.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')
grafico.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')

grafico
````

Este modelo deve prever a varivel ```renda``` com base nas demais variveis exceto ```data_ref``` e ```index```. J vimos que a varivel renda  melhor modelada com a transformao ```log()```, no se esquea disso.

````python
import pandas as pd
import matplotlib as plt
%matplotlib inline
import numpy as np
import statsmodels.formula.api as smf
import statsmodels.api as sm
import seaborn as sns
from sklearn.model_selection import train_test_split
import scipy.stats as ss 
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
from sklearn.metrics import r2_score 
````

````python
df = pd.read_csv('previsao_de_renda_II.csv')
df.head()
````

````python
df.info()
````

````python
df['tempo_emprego'] = df['tempo_emprego'].fillna(df['tempo_emprego'].mean())
````

````python
df['data_ref'] = pd.to_datetime(df['data_ref'])
df['mes'] = df['data_ref'].dt.month
````

````python
train, test = train_test_split(df, random_state=0, train_size = .7)
````

````python
reg = smf.ols('''np.log(renda) ~ sexo + posse_de_veiculo + posse_de_imovel + qtd_filhos + tipo_renda + 
                educacao + estado_civil + idade + tipo_residencia + tempo_emprego + qt_pessoas_residencia''',data = train).fit()
reg.summary()
````

## Outliers

Avalie os *studentized residuals*, verifique se h pontos que parecem ser discrepantes.

Avalie se h pontos influentes.

````python
train_sample = train.head(525)
````

````python
reg_sample = smf.ols('''np.log(renda) ~ sexo + posse_de_veiculo + posse_de_imovel + qtd_filhos + tipo_renda + 
                        educacao + estado_civil + idade + tipo_residencia + tempo_emprego + qt_pessoas_residencia''',
                     data = train_sample).fit()
reg_sample.summary()
````

````python
st_res = reg_sample.outlier_test()
sns.scatterplot(x = reg_sample.predict(), y = st_res.student_resid);
````

````python
fig = sm.graphics.influence_plot(reg_sample, criterion="cooks", size=12)
````

A partir das anlise de outliers e pontos influentes, percebe-se que h alguns pontos que podem estar influenciando negativamente o modelo.

## Multicolinearidade

Avalie se h questes relacionadas a multicolinearidade atravs de pelo menos:

- Matriz de correlao de Spearman
- VIF

````python
train[['qtd_filhos','idade','tempo_emprego','qt_pessoas_residencia']].corr(method='spearman')
````

````python
variaveis = ['sexo','posse_de_veiculo','posse_de_imovel','qtd_filhos','tipo_renda','educacao','estado_civil',
            'tipo_residencia','idade','tempo_emprego','qt_pessoas_residencia']
train_dum = pd.get_dummies(train[variaveis], drop_first=True)
````

````python
vars_vif = pd.DataFrame()
vars_vif["VIF Factor"] = [vif(train_dum, i) for i in range(train_dum.shape[1])]
vars_vif["Feature"] = train_dum.columns
vars_vif.round(2)
````

As anlises acima apresentam algumas informaes relevantes sobre a importncia de algumas variveis no modelo e a correlao entre elas, exemplo a grande correlao entre 'qtd_filhos' e 'qt_pessoas_residencia', alm da pouca relevncia de 'qt_pessoas_residencia' no modelo.

## Ajustes

Faa os ajustes que julgar necessrios no modelo e compare as mtricas de desempenho do modelo original e ajustado na base de testes.

````python
reg_ajustado = smf.ols('''np.log(renda) ~ sexo + posse_de_veiculo + posse_de_imovel + tipo_renda + 
                        tempo_emprego''',data = train).fit()
reg_ajustado.summary()
````

````python
test['preditos'] = reg.predict(test.loc[:,test.columns!='renda'])
test['preditos_ajustado'] = reg_ajustado.predict(test.loc[:,test.columns!='renda'])
````

````python
adj_r2 = (1-((1-r2_score(test['renda'],test['preditos']))*(len(test['preditos'])-1))/
          (len(test['preditos']) - test.shape[1]-1))
print("A acurcia foi de {:.2f}%.".format(adj_r2))
````

````python
adj_r2 = (1-((1-r2_score(test['renda'],test['preditos_ajustado']))*(len(test['preditos_ajustado'])-1))/
          (len(test['preditos_ajustado']) - test.shape[1]-1))
print("A acurcia foi de {:.2f}%.".format(adj_r2))
````

Apesar dos ajustes, o resultado obtido no indica melhora no modelo.
